```python
import numpy as np

# Define a 3d one-to-one function
def f(x, y, z):
  return x * y + z

# Create a NumPy array of input values
x = np.linspace(1, 5, 5)
y = np.linspace(1, 5, 5)
z = np.linspace(1, 5, 5)

# Apply the function to the input values
f_values = f(x, y, z)

# Create a 3d NumPy array of the function outputs
f_array = np.stack((x, y, f_values), axis=2)

# Plot the function
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.plot_surface(f_array)

# Display the diagram
plt.show()
```
