```python
import numpy as np
import matplotlib.pyplot as plt

years = np.array([2010, 2012, 2014, 2016, 2018, 2020, 2022])
users = np.array([100, 300, 500, 700, 1200, 2000, 3000])

plt.figure(figsize=(5, 5))
plt.plot(years, users, marker='o')
plt.xlabel('Years')
plt.ylabel('Number of Users/Transactions')
plt.title('Blockchain Technology Growth')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate longitude and latitude arrays
lon = np.linspace(-180, 180, 360)
lat = np.linspace(-90, 90, 180)

# Create a meshgrid of longitude and latitude coordinates
lon_grid, lat_grid = np.meshgrid(lon, lat)

# Set a constant value for the entire grid
data = np.ones_like(lon_grid)

plt.figure(figsize=(10, 6))
plt.contourf(lon_grid, lat_grid, data, cmap='terrain')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('2D Diagram of Earth')
plt.colorbar()
plt.grid(True)
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
# Generate longitude and latitude arrays
lon = np.linspace(-np.pi, np.pi, 360)
lat = np.linspace(-np.pi / 2, np.pi / 2, 180)

# Create a meshgrid of longitude and latitude coordinates
lon_grid, lat_grid = np.meshgrid(lon, lat)

# Set a constant value for the entire grid
data = np.ones_like(lon_grid)
plt.figure(figsize=(8, 8))
plt.pcolormesh(lon_grid, lat_grid, data, cmap='terrain', shading='auto')
plt.axis('equal')
plt.axis('off')
plt.title('Circular Graph of Earth')
plt.colorbar()
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
continent_names = ['Africa', 'Antarctica', 'Asia', 'Europe', 'North America', 'Oceania', 'South America']
continent_areas = [30, 5, 44, 10, 17, 8, 12]  # Sample areas in million square kilometers
total_area = sum(continent_areas)
angles = [area / total_area * 2 * np.pi for area in continent_areas]
plt.figure(figsize=(3, 3))
plt.pie(angles, labels=continent_names, startangle=50, counterclock=False)
plt.title('Circular Graph of Continents')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

map = Basemap(projection='mill', llcrnrlat=-90, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180, resolution='c')
map.drawcoastlines()
map.drawcountries()
map.drawmapboundary(fill_color='lightblue')
plt.title('World Map')
plt.show()
```
