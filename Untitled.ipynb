```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D array using NumPy
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

# Create a new figure or activate an existing figure using the figure() method
fig = plt.figure()

# Add an Axes to the figure as part of a subplot arrangement using add_subplot() method
ax = fig.add_subplot(111, projection='3d')

# Use the plot_surface() method to plot the 3D graph
ax.plot_surface(x, y, z)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Frequency range of visible light colors
frequencies = {
    'Red': [400, 484],
    'Orange': [484, 508],
    'Yellow': [508, 526],
    'Green': [526, 606],
    'Blue': [606, 668],
    'Indigo': [668, 688],
    'Violet': [688, 789]
}

# Plot the frequency ranges
fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
theta = np.linspace(0, 2 * np.pi, len(frequencies))
radii = np.array([freq[1] - freq[0] for freq in frequencies.values()])
width = np.pi / 4 * np.ones(len(frequencies))
bars = ax.bar(theta, radii, width=width, bottom=400)
for bar, color in zip(bars, frequencies.keys()):
    bar.set_facecolor(color.lower())
    bar.set_alpha(0.5)

plt.title('Frequency of Visible Light Colors')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation

# Load the Earth texture
img = plt.imread('earth.png')

# Create a sphere
r = 1
pi = np.pi
cos = np.cos
sin = np.sin
phi, theta = np.mgrid[0.0:pi:100j, 0.0:2.0*pi:100j]
x = r*sin(phi)*cos(theta)
y = r*sin(phi)*sin(theta)
z = r*cos(phi)

# Set up the figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the Earth
earth = ax.plot_surface(
    x, y, z, rstride=1, cstride=1, facecolors=img)

# Set the axis limits and remove the axis labels
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([-1.5, 1.5])
ax.axis('off')

# Function to rotate the Earth by one degree
def rotate(i):
    ax.view_init(elev=10., azim=i)

# Create an animation of the rotating Earth
ani = animation.FuncAnimation(fig, rotate, frames=360, interval=20)
plt.show()
```

```python
# Server
import socket
import numpy as np

# Create a socket object
s = socket.socket()

# Get local machine name
host = socket.gethostname()

# Reserve a port for your service
port = 12345

# Bind to the port
s.bind((host, port))

# Wait for client connection
s.listen(5)

# Accept a connection
c, addr = s.accept()
print('Got connection from', addr)

# Receive the data in small chunks and reassemble it
data = b''
while True:
    # Receive 16 bytes of data
    packet = c.recv(16)
    if not packet: 
        break
    # Append the data to the buffer
    data += packet

# Convert the bytes data to numpy array
arr = np.frombuffer(data, dtype=np.uint8)
print(arr)

# Close the connection
c.close()
```

```python
# Client
import socket
import numpy as np

# Create a socket object
s = socket.socket()

# Get local machine name
host = socket.gethostname()

# Reserve a port for your service
port = 12345

# Connect to the server
s.connect((host, port))

# Create a random numpy array of uint8 type
arr = np.random.randint(0, 255, size=100, dtype=np.uint8)
print(arr)

# Convert the numpy array to bytes
data = arr.tobytes()

# Send the data to the server
s.sendall(data)

# Close the socket
s.close()
```
